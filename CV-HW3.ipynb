```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from google.colab import drive
import os
```

```python
os.chdir('/content/drive/MyDrive/CV_HW3')

img1 = cv2.imread('3.jpg')
img2 = cv2.imread('8.jpg')
img3 = cv2.imread('5.jpg')
img4 = cv2.imread('6.jpg')
img5 = cv2.imread('10.jpg')

images = [img1, img2, img3, img4, img5]

fig = plt.figure(figsize=(20, 10))

for i, img in enumerate(images):
    plt.subplot(1, len(images), i + 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f'Image {i+1}')

plt.show()
```

```python
def plot_lines(img,edges,th,mll,mlg):

  lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=th, minLineLength=mll, maxLineGap=mlg)

  output = img.copy()

  if lines is not None:
    for line in lines:
      x1, y1, x2, y2 = map(int, line[0])
      cv2.line(output, (x1, y1), (x2, y2), (0, 255, 0), 2)

  plt.figure(figsize=(15, 10))
  plt.subplot(131), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  plt.title('Original Image'), plt.axis('off')
  plt.subplot(132), plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
  plt.title('Edges'), plt.axis('off')
  plt.subplot(133), plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
  plt.title('Line Detection'), plt.axis('off')
  plt.show()
```

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

blurred1 = cv2.GaussianBlur(gray1, (9, 9), 0)

edges1 = cv2.Canny(blurred1, 50, 150, apertureSize=3)

kernel = np.ones((3, 3), np.uint8)
dilated1=cv2.dilate(edges1,kernel,iterations=1)


plot_lines(img1,dilated1,th=80,mll=50,mlg=50)
```
